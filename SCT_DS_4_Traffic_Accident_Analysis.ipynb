# SCT_DS_4 - Traffic Accident Analysis

## Objective

The objective of this project is to analyze traffic accident data and identify patterns related to accident severity, weather conditions, road surface conditions, lighting conditions, speed limits, and area types. Through exploratory data analysis and visualization, the project aims to uncover key factors that contribute to road accidents and provide insights that can support road safety improvements.

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Accident_Information.csv")
```

```python
df.sample(5)
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(
    columns=[
        '2nd_Road_Class',
        'Carriageway_Hazards',
        'Special_Conditions_at_Site'
    ],
    inplace=True
)
```

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Accident_Severity',
    data=df,
    palette='pastel'
)

plt.title('Distribution of Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()
```

Observation

Slight accidents account for the majority of recorded incidents.

Serious accidents occur less frequently, while fatal accidents represent the smallest proportion.

The distribution indicates that most road accidents result in non-fatal outcomes.

```python
top_weather = df['Weather_Conditions'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_weather.values,
    y=top_weather.index,
    palette='pastel'
)

plt.title('Top Weather Conditions Associated with Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.show()
```

Observation

Certain weather conditions are associated with a higher number of accidents.

Adverse weather can reduce visibility and vehicle control, increasing accident risk.

Weather remains an important external factor influencing road safety.

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='Light_Conditions',
    data=df,
    palette='pastel'
)

plt.title('Accidents by Light Conditions')
plt.xlabel('Count')
plt.ylabel('Light Conditions')
plt.show()
```

Observation

Daylight conditions account for a significant share of reported accidents.

Accidents occurring during darkness highlight the impact of reduced visibility on road safety.

Lighting conditions play an important role in accident occurrence and severity.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    y='Road_Surface_Conditions',
    data=df,
    palette='pastel'
)

plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Count')
plt.ylabel('Road Surface Condition')
plt.show()
```

Observation

Dry road surfaces account for the highest number of accidents, likely due to higher traffic volume.

Wet or slippery road conditions continue to contribute significantly to accident occurrences.

Road surface quality remains an important factor affecting driver safety.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Speed_limit',
    data=df,
    palette='pastel'
)

plt.title('Accidents Across Different Speed Limits')
plt.xlabel('Speed Limit')
plt.ylabel('Count')
plt.show()
```

Observation

Accident frequency varies across different speed limits.

Roads with moderate speed limits often experience the highest number of incidents due to increased traffic flow.

Speed regulation plays a crucial role in reducing accident severity.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Urban_or_Rural_Area',
    data=df,
    palette='pastel'
)

plt.title('Urban vs Rural Accident Distribution')
plt.xlabel('Area Type')
plt.ylabel('Count')
plt.show()
```

Observation

Urban areas generally record a higher number of accidents compared to rural regions.

Increased traffic density and road usage contribute to accident frequency in urban environments.

Rural accidents, although fewer, may involve higher severity due to road characteristics and speed limits.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Number_of_Casualties'],
    bins=20,
    color='#A8DADC'
)

plt.title('Distribution of Number of Casualties')
plt.xlabel('Number of Casualties')
plt.ylabel('Frequency')
plt.show()
```

Observation

Most accidents involve a relatively small number of casualties.

High-casualty accidents are comparatively rare.

The distribution suggests that severe multi-casualty incidents occur less frequently than minor accidents.

```python
severity_map = {
    'Slight': 1,
    'Serious': 2,
    'Fatal': 3
}

df['Severity_Numeric'] = df['Accident_Severity'].map(severity_map)
```

```python
top_weather = df['Weather_Conditions'].value_counts().head(5).index

filtered_df = df[df['Weather_Conditions'].isin(top_weather)]

plt.figure(figsize=(12,6))

sns.countplot(
    x='Accident_Severity',
    hue='Weather_Conditions',
    data=filtered_df,
    palette='pastel'
)

plt.title('Accident Severity by Weather Condition')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.legend(title='Weather Conditions')
plt.show()
```

Observation

The distribution of accident severity varies across different weather conditions.

Clear weather conditions account for the highest number of recorded accidents, likely due to higher traffic volume and road usage.

Adverse weather conditions such as rain, fog, and strong winds are associated with a noticeable proportion of serious accidents.

The findings suggest that weather conditions can influence both the occurrence and severity of road accidents, highlighting the importance of cautious driving during unfavorable weather.
